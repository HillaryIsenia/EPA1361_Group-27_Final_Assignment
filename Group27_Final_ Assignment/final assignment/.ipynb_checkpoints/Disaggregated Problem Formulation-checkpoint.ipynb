````python
#validate with different random.seed experiments
def extra_trees(seed):
    random.seed(seed)
    with MultiprocessingEvaluator(dike_model) as evaluator:
        results = evaluator.perform_experiments(scenarios=1000, policies=policy_0)
    experiments, outcomes = results
    cleaned_experiments = experiments.drop(columns=[l.name for l in dike_model.levers])
    for outcome in outcomes: #aanpassen op uitkomsten dict
        outcome_list = outcomes[outcome] #aanpassen op uitkomsten dict
        scores = feature_scoring.get_ex_feature_scores(cleaned_experiments, outcome_list, nr_trees = 1000, max_features=0.6, mode=RuleInductionType.REGRESSION)
        if outcome == 'Gelderland Expected Annual Damage': #niet nodig
            Gelderland_Damage_scores = scores #niet nodig
            Gelderland_Damage_scores = Gelderland_Damage_scores[0]
            GEAD = Gelderland_Damage_scores.T
            GEAD = GEAD.rename(index={1: 'Gelderland Expected Annual Damage'})
        elif outcome == 'Gelderland Expected Number of Deaths':
            Gelderland_Death_scores = scores
            Gelderland_Death_scores = Gelderland_Death_scores[0]
            GEND = Gelderland_Death_scores.T
            GEND = GEND.rename(index={1: 'Gelderland Expected Number of Deaths'})
        elif outcome == 'Overijssel Expected Annual Damage': 
            Overijssel_Damage_scores = scores #niet nodig
            Overijssel_Damage_scores = Overijssel_Damage_scores[0]
            OEAD = Overijssel_Damage_scores.T
            OEAD = OEAD.rename(index={1: 'Overijssel Expected Annual Damage'})
        else: #niet nodig
            Overijssel_Death_scores = scores #niet nodig
            Overijssel_Death_scores = Overijssel_Death_scores[0]
            OEND = Overijssel_Death_scores.T
            OEND = OEND.rename(index={1: 'Overijssel Expected Number of Deaths'})
    fs_sa = GEND.append(GEAD)
    fs_sa = fs_sa.append(OEAD)
    fs_sa = fs_sa.append(OEND)
    fig = plt.figure(figsize =(18, 10))

    sns.heatmap(fs_sa.T, cmap='viridis', annot=True)

    plt.show()
    return fs_sa
````

````python
def apply_prim(box_number, y = GEAD, threshold = 0.8, peel_alpha=0.05):
    prim_alg = prim.Prim(x = cleaned_experiments, y=y, threshold=threshold, peel_alpha=peel_alpha)
    box = prim_alg.find_box() 
    
    #Visualize the trade off between coverage and density
    box.show_tradeoff()
    plot1 = plt.show()
    plot2 = box.inspect(style='graph')
    plot2.set_size_inches(18,18)
    
    #Plot the point of box 2 in a pairs_scatter plot
    box.select(box_number)
    fig = box.show_pairs_scatter()
    plot3 = plt.show()
    return (plot1, plot2) # plot3)
````

## 5. Multi-Objective Robust Optimization (MORO)
In this part of the notebook, the candidate policy alternatives are identified using MORO.

````python
def robustness(direction, threshold, data):
    if direction == SMALLER:
        return np.sum(data<=threshold)/data.shape[0]
    else:
        return np.sum(data>=threshold)/data.shape[0]

SMALLER = 'SMALLER'
LARGER = 'LARGER'

G_Annual_Damage = functools.partial(robustness, SMALLER, 1618855)
G_Number_Deaths = functools.partial(robustness, SMALLER, 1)
O_Annual_Damage = functools.partial(robustness, SMALLER, 236987)
O_Number_Deaths = functools.partial(robustness, SMALLER, 1)
````

The code of the experimental setup for the robustness converges in pseudo code is:

```python
from ema_workbench import (MultiprocessingEvaluator, save_results)
from ema_workbench.em_framework.evaluators import perform_experiments

random.seed(3922875)
with MultiprocessingEvaluator(dike_model) as evaluator:
    results = evaluator.perform_experiments(scenarios=500,               
                                            policies=4, uncertainty_sampling='mc')

save_results((results), './results/pf7_experiments.tar.gz')
```

````python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(15,15),
                        sharex=True)
axes = [axes[0,0],axes[0,1],axes[1,0],axes[1,1]]                             #axes[1,1]

robustness_funcs = {'Gelderland Expected Annual Damage': G_Annual_Damage,
                    'Gelderland Expected Number of Deaths': G_Number_Deaths,
                    'Overijssel Expected Annual Damage' : O_Annual_Damage,
                    'Overijssel Expected Number of Deaths' : O_Number_Deaths}

experiments, outcomes = scenario_check_results

for ax, (outcome, value) in zip(axes, outcomes.items()):
    for policy in np.unique(experiments['policy']):
        logical = experiments['policy'] == policy
        data = value[logical]
        
        robustness = []
        #print(outcome)
        for i in range(1, data.shape[0]):
            robustness.append(robustness_funcs[outcome](data[0:i]))
            #print(data)
        ax.plot(robustness, label=policy)
    ax.set_xlabel("# experiments")
    ax.set_ylabel(outcome)
````
